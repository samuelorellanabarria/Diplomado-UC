# Graficos de Barra
- **Grafico de barras** : Utilizado para representar un atributo categorico y otro numerico. Normalmente el eje "X" es categorico y el eje "Y" es numerico. Es un tipo de grafico muy preciso para hacer comparaciones
- **Grafico de barras apiladas**: Permite codificar visualmente 2 variables categóricas, y una numérica

```python
# Grafico Matplotlib Base
categorias = ['A', 'B', 'C', 'D', 'E']
valores = [10, 20, 15, 25, 30]
data = {'Categorías': categorias, 'Valores': valores}
df = pd.DataFrame(data)

plt.bar(df['Categorías'], df['Valores']) # mas poder de personalización, viene por matplotlib
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.title('Gráfico de Barra')
plt.show()

#df.plot(kind='bar', x='Categorías', y='Valores') # graficos simples, trabaja con pandas
#plt.show()
```

```python
# Grafico Seaborn Base
categorias = ['A', 'B', 'C', 'D', 'E']
valores = [10, 20, 15, 25, 30]
data = {'Categorías': categorias, 'Valores': valores}
df = pd.DataFrame(data)

sns.barplot(x='Categorías', y='Valores', data=df)
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.title('Gráfico de Barra')
plt.show()
```

```python
# Grafico Matplotlib con Groupby
# Datos de ejemplo
data = {
    'Categoría': ['A', 'A', 'B', 'B', 'C', 'C', 'D', 'D'],
    'Mes': ['Enero', 'Febrero', 'Enero', 'Febrero', 'Enero', 'Febrero', 'Enero', 'Febrero'],
    'Ventas': [100, 150, 200, 220, 150, 170, 300, 330]
}
df = pd.DataFrame(data)

# Agrupar los datos por categoría y sumar las ventas
grouped_df = df.groupby('Categoría')['Ventas'].sum().reset_index()

# Crear el gráfico de barras
plt.bar(grouped_df['Categoría'], grouped_df['Ventas'])

# Etiquetas y título
plt.xlabel('Categoría')
plt.ylabel('Ventas Totales')
plt.title('Ventas Totales por Categoría')

# Mostrar el gráfico
plt.show()
```

```python
# Grafico Seaborn con Groupby
# Datos de ejemplo
data = {
    'Categoría': ['A', 'A', 'B', 'B', 'C', 'C', 'D', 'D'],
    'Mes': ['Enero', 'Febrero', 'Enero', 'Febrero', 'Enero', 'Febrero', 'Enero', 'Febrero'],
    'Ventas': [100, 150, 200, 220, 150, 170, 300, 330]
}
df = pd.DataFrame(data)

# Crear el gráfico de barras con agregación
sns.barplot(x='Categoría', y='Ventas', data=df, estimator=sum, ci=None)

# Etiquetas y título
plt.xlabel('Categoría')
plt.ylabel('Ventas Totales')
plt.title('Ventas Totales por Categoría')

# Mostrar el gráfico
plt.show()
```

```python
# Graficos de barras apiladas con Matplotlib
# Datos de ejemplo
data = {
    'Categoría': ['A', 'A', 'B', 'B', 'C', 'C', 'D', 'D'],
    'Mes': ['Enero', 'Febrero', 'Enero', 'Febrero', 'Enero', 'Febrero', 'Enero', 'Febrero'],
    'Ventas': [100, 150, 200, 220, 150, 170, 300, 330]
}
df = pd.DataFrame(data)

# Agrupar los datos por categoría y mes
grouped_df = df.groupby(['Categoría', 'Mes'])['Ventas'].sum().unstack().fillna(0)

# Crear el gráfico de barras apiladas
grouped_df.plot(kind='bar', stacked=True, figsize=(10, 6))

# Etiquetas y título
plt.xlabel('Categoría')
plt.ylabel('Ventas')
plt.title('Ventas por Categoría y Mes (Apiladas)')
plt.legend(title='Mes')

# Mostrar el gráfico
plt.show()
```

# Graficos de Puntos y Lineas
- **Grafico de Puntos**: Utilizados principalmente para presentar tendencias // En el eje X se pone una variable ordinal (típicamente datos temporales)
- **Grafico de Lineas**: El gráfico de líneas es simular al de puntos, sin embargo, las lineas que unen los puntos permiten dar una sensación de continuidad // En el eje X pueden perfectamente ponerse cientos de niveles (valores) y el gráfico escala bien visualmente
- **No confunfir** el gráfico de puntos con el grafico de dispersión, donde tanto el eje X como el eje Y presentan valores de atributos numéricos.

**MATPLOTLIB**

```python
# Creación del DataFrame
data = {
    'Categoría': ['A', 'B', 'C', 'D', 'E', "F", "G", "H", "I", "J", "K", "L", "O", "P", "Q"],
    'Valor': [1, 2, 3, 3, 2, 3, 5, 4, 5, 3, 4, 2, 5, 3, 5]
}
df = pd.DataFrame(data)

# Crear gráfico de puntos con Matplotlib
plt.plot(df['Categoría'], df['Valor'], marker='o', linestyle='None')
plt.title('Gráfico de Puntos con Matplotlib')
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.show()
```

```python
# Creación del DataFrame
data = {
    'Categoría': ['A', 'B', 'C', 'D', 'E', "F", "G", "H", "I", "J", "K", "L", "O", "P", "Q"],
    'Valor': [1, 2, 3, 3, 2, 3, 5, 4, 5, 3, 4, 2, 5, 3, 5]
}
df = pd.DataFrame(data)

# Crear gráfico de puntos con Matplotlib
plt.plot(df['Categoría'], df['Valor'], marker='o', linestyle='-')
plt.title('Gráfico de Puntos con Matplotlib')
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.show()
```

**SEABORN**
**Características de seaborn.lineplot**

- **x, y**: Las variables para los ejes x e y.
- **data**: El DataFrame que contiene los datos.
- **hue**: Variable para agrupar y colorear las líneas.
- **style**: Variable para aplicar diferentes estilos de línea.
- **markers**: Parámetro para agregar marcadores a las líneas.
- **err_style**: Forma de representar el error ("band" o "bars").
- **ci**: Intervalo de confianza (por defecto al 95%).

```python
# Creación del DataFrame
data = {
    'Categoría': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K', 'L', 'O', 'P', 'Q'],
    'Valor': [1, 2, 3, 3, 2, 3, 5, 4, 5, 3, 4, 2, 5, 3, 5]
}
df = pd.DataFrame(data)

# Crear gráfico de líneas con puntos usando Seaborn
sns.lineplot(x='Categoría', y='Valor', data=df, marker='o', linewidth=0) #linewidth=0
plt.title('Gráfico de Líneas con Puntos con Seaborn')
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.show()
```

```python
# Creación del DataFrame
data = {
    'Categoría': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K', 'L', 'O', 'P', 'Q'],
    'Valor': [1, 2, 3, 3, 2, 3, 5, 4, 5, 3, 4, 2, 5, 3, 5]
}
df = pd.DataFrame(data)

# Crear gráfico de líneas con puntos usando Seaborn
sns.lineplot(x='Categoría', y='Valor', data=df, marker='o', color="red", linewidth=1, err_style="band")
plt.title('Gráfico de Líneas con Puntos con Seaborn')
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.show()
```

**Diferencias Clave**
1. **Método 1**: Sin asignar el objeto del eje
- Simplicidad: Es directo y sencillo, ideal para gráficos rápidos.
- Configuración global: Usa funciones globales de matplotlib (plt.title, plt.xlabel, plt.ylabel), lo que es suficiente para gráficos simples.
2. **Método 2**: Asignando el objeto del eje (ax)
- Mayor control: Asigna el gráfico a una variable ax, lo que permite personalizar y manipular el gráfico de manera más detallada.
- Configuración específica: Puedes usar métodos específicos del objeto ax para ajustar títulos, etiquetas y otros atributos, proporcionando mayor flexibilidad.
- Ideal para subplots: Es muy útil cuando trabajas con múltiples subplots (gráficos en la misma figura), ya que puedes personalizar cada gráfico de manera independiente.

**Ejemplo de Uso Avanzado con el Objeto ax**
Cuando necesitas trabajar con múltiples gráficos en la misma figura, usar el objeto ax es esencial:

```python
data = {
    'Categoría': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K', 'L', 'O', 'P', 'Q'],
    'Valor': [1, 2, 3, 3, 2, 3, 5, 4, 5, 3, 4, 2, 5, 3, 5]
}
df = pd.DataFrame(data)

fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Primer gráfico de líneas
sns.lineplot(ax=axes[0], x='Categoría', y='Valor', data=df, marker='o', color="blue", linewidth=1, err_style="band")
axes[0].set_title('Gráfico de Líneas 1')
axes[0].set_xlabel('Categorías')
axes[0].set_ylabel('Valores')

# Segundo gráfico de líneas
sns.lineplot(ax=axes[1], x='Categoría', y='Valor', data=df, marker='o', color="green", linewidth=1, err_style="band")
axes[1].set_title('Gráfico de Líneas 2')
axes[1].set_xlabel('Categorías')
axes[1].set_ylabel('Valores')

plt.tight_layout()
plt.show()
```

# Graficos de Dispersión y Burbujas
- **Grafico de Dispersión**: Utilizado para visualizar relaciones entre 2 atributos numericos // Permite visualizar fácilmente si es que existe correlación, tanto positiva como negativa // En ingles se conoce como *scatterplots*
- **Gráfico de Burbujas**: Es la Extensión del gráfico de Dispersión // Permite codificar variables categóricas adicionales usando los canales de color y tamaño de los puntos

```python
data = {
    'x': [1, 2, 3, 4, 5],
    'y': [2, 3, 5, 7, 11],
    'sizes': [20, 50, 100, 200, 500],  # Tamaño de las burbujas
    'colors': [1, 2, 3, 4, 5]  # Color de las burbujas
}

# Crear DataFrame
df = pd.DataFrame(data)
```

**MATPLOTLIB**

```python
# Crear el gráfico de dispersión
plt.scatter(df['x'], df['y'], color='blue')
plt.title('Gráfico de Dispersión con matplotlib')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.show()
```

```python
# Crear el gráfico de burbujas
plt.scatter(df['x'], df['y'], s=df['sizes'], c=df['colors'], alpha=0.5, cmap='viridis')
plt.title('Gráfico de Burbujas con matplotlib')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.colorbar()  # Mostrar barra de colores
plt.show()
```

**SEABORN**

```python
# Crear el gráfico de dispersión
sns.scatterplot(x='x', y='y', data=df)
plt.title('Gráfico de Dispersión con seaborn')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.show()
```

```python
# Crear el gráfico de burbujas
sns.scatterplot(x='x', y='y', size='sizes', hue='colors', data=df, alpha=0.5, palette='viridis')
plt.title('Gráfico de Burbujas con seaborn')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend(title='Colors')
plt.show()
```

# Graficos de Flujo
- **Grafico de Flujo**: Utilizado para visualizar tendencias en el tiempo // A diferencia del gráfico de líneas, usa el canal de área y de color para presentar la información de tendencias. // En ingles se conoce como *Streamgraph*
- **Se codifican visualmente 3 variables**:
-- Un atributo temporal (eje X),
-- Un atributo numérico para cuantificar el tamaño de las áreas,
-- Y un atributo categórico para mostrar las distintas áreas, normalmente codificadas con color
- Se requiere la libreria **Altair as alt**

```python
import altair as alt
```

```python
# Luego le indicamos a altair que estaremos trabajando en un notebook.
alt.renderers.enable('default')
```

```python
# Definir géneros y años
genres = ['acción', 'comedia', 'drama', 'terror', 'ciencia ficción']
years = np.arange(2010, 2021)

# Generar datos aleatorios para cada combinación de género y año
data = {
    'genre': np.repeat(genres, len(years)),
    'year': np.tile(years, len(genres)),
    'count': np.random.randint(5, 25, size=len(genres) * len(years))
}

# Crear el DataFrame
df = pd.DataFrame(data)
print(df)
```

# Histograma, Gráfico de caja, gráfico de violin
- **Histograma**: Permite conocer la distribución de los valores de un atributo numérico (cuantitativo y continuo) // ejemplo: la distribución de edades de un grupo de personas
- **Gráfico de Caja**: Permite visualizar la distribución de datos de una variable numérica (cuantitativa continua), enfatizando medidas estafísticas como la mediana, cuartiles y outliers. En ingles se denomina **Box plot**. // La caja central está determinada por el cuartil Q1 (percentil 25) y el cuartil Q3 (percentil 75). La línea dentro de la caja indica la mediana (percentil 50). // La media aritmética no siempre se incluye en el gráfico de caja. cuando se incluye, se usa una marca adicional (un asterisco, por ejemplo) // Rango intercuantil (Q25, Q75)
- **Gráfico de violín**: Similar al gráfico de caja en cuanto a su propósito: presentar información resumida visualmente, de una distribución de datos // Diferenciación del gráfico de caja: por que al usar un pre-cálculo de la distribución de probabilidad estimada, dibuja una forma suavisada de la distribución directamente desde los datos. // Es más efectivo que el gráfico de caja en términos visuales: pero como la forma suavisada es aproximada, puede contener sesgos si hay muy pocos datos.

**sns.set()**  # Aplica configuraciones predeterminadas de Seaborn

1. Cuando llamas a sns.set(), Seaborn configura:
- Estilo de las parcelas: whitegrid (predeterminado).
- Paleta de colores: deep (predeterminado).
- Contexto: notebook (predeterminado).
- Fuente: sans-serif.

Diferencias entre las funciones de histogramas
- sns.distplot: Obsoleto. Combina histograma y KDE en un solo gráfico.
- sns.histplot: Actual. Crea histogramas avanzados con opciones para personalizar KDE.
- sns.displot: Actual. Función de alto nivel para distribuciones, permite facetado y múltiples tipos de gráficos (histogramas, KDE, ECDF).

```python
# Definir la semilla para reproducibilidad
np.random.seed(0)

# Número de filas deseadas
num_rows = 100

# Posibles valores para cada columna categórica
sexes = ['Male', 'Female']
smokers = ['Yes', 'No']
days = ['Thur', 'Fri', 'Sat', 'Sun']
times = ['Lunch', 'Dinner']
sizes = [1, 2, 3, 4, 5, 6]

# Generar datos aleatorios
extended_data = {
    'total_bill': np.random.uniform(10, 50, num_rows).round(2),
    'tip': np.random.uniform(1, 10, num_rows).round(2),
    'sex': np.random.choice(sexes, num_rows),
    'smoker': np.random.choice(smokers, num_rows),
    'day': np.random.choice(days, num_rows),
    'time': np.random.choice(times, num_rows),
    'size': np.random.choice(sizes, num_rows)
}

# Crear el DataFrame
df = pd.DataFrame(extended_data)
df.head()
```

**HISTOGRAMA**

```python
ax = sns.histplot(df.tip, kde=True, bins=20)
```

```python
sns.set()
ax = sns.histplot(df.tip, kde=False, color="red", bins=20)
```

```python
# VARIOS HISTOGRAMAS EN UN MISMO GRAFICO
sns.histplot(df[df.time == "Lunch"].tip, kde=False,
            color="red", label="Propinas de Almuerzo", bins=25)
sns.histplot(df[df.time == "Dinner"].tip, kde=False,
            color="skyblue", label="Propinas de Cena", bins=25)
plt.legend()
plt.show()
```
